```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def load_error_seq(filename):
    error_seq = [ float(s.strip()) for s in open(filename) ]
    return range(len(error_seq)),error_seq
```

```python
plt.figure(figsize=(20,10))
X1,Y1 = load_error_seq('train.standard.log')
plt.plot(X1,Y1)
X2,Y2 = load_error_seq('train.standard.smart_init.log')
plt.plot(X2,Y2)
```
